```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.preprocessing import normalize 
%matplotlib inline
```

```python
numericalDF = pd.read_csv('numericalData.csv')
columnas = ['directores','criticas','duracion','genero','actores1','actores2','actores3','rostros',
            'scores','gross','budget']
del numericalDF['Unnamed: 0']

for i in columnas:
    numericalDF.loc[numericalDF[i] == 0, i] = numericalDF[i].mean()
X = numericalDF.head(20)
numericalDF.head()
```

empezaremos con los metodos de clustering 
el primero sera la distancia, aqui podremos tomar una primera aproximacion de que tan lejos esta un punto de otro, podemos considerar esto como un acercamiento a la primera forma en la que podemos ver que tipo de clusters se encuentran o mejor dicho cuantos se encuentran
a continuacion veremos distintas formas de distancia, que son euclideana, manhattan y minkowski

```python
from scipy.spatial import distance
distanciasEuclidians = distance.pdist(X,"euclidean")
print('\nLas distancias euclidianas ({}) para los datos de peliculas son:\n'.format(distanciasEuclidians.size), distanciasEuclidians)

ms = 1/distanciasEuclidians
print('\nLas similaridades euclidianas para los datos de peliculas son:\n', ms)

md2 = distanciasEuclidians/10000
print('\nLas similaridades euclidianas para los datos "pequeños" de peliculas son:\n', 1/md2)
```

```python
distanciasManhattan = distance.pdist(X,'cityblock')
print('\nLas distancias Manhattan ({}) para los datos de peliculas son:\n'.format(distanciasManhattan.size), distanciasManhattan)

ms = 1/distanciasManhattan
print('\nLas similaridades menhattan para los datos de peliculas son:\n', ms)

md2 = distanciasManhattan/10000
print('\nLas similaridades manhattan para los datos "pequeños" de peliculas son:\n', 1/md2)
```

```python
distanciasMinkowski = distance.pdist(X, 'minkowski', 1)
print('\nLas distancias Minkowski ({}) para los datos de peliculas son:\n'.format(distanciasMinkowski.size), distanciasMinkowski)

ms = 1/distanciasMinkowski
print('\nLas similaridades minkowski para los datos de peliculas son:\n', ms)

md2 = distanciasMinkowski/10000
print('\nLas similaridades minkowski para los datos "pequeños" de peliculas son:\n', 1/md2)
```
