```python
import pandas as pd
import numpy as np
import matplotlib as plt
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
%matplotlib inline
```

```python
numericalDF = pd.read_csv('numericalData.csv')
columnas = ['directores','criticas','duracion','genero','actores1','actores2','actores3','rostros',
            'scores','gross','budget','class']
del numericalDF['Unnamed: 0']
del numericalDF['class']
for i in columnas:
    numericalDF.loc[numericalDF[i] == 0, i] = numericalDF[i].mean()
X = numericalDF.head(20)
numericalDF.head()
```

iremos al segundo metodo de clustering, que sera la utlizacion de dendogramas, con esto podremos identificar el numero de clusters que hay en los datos, facilmente identificables, porque los ordena en una forma jerarquica que lo hace facil de encontrar
aqui demostraremos distintos tipos

```python
Z = linkage(X, 'single')
plt.figure(figsize=(12,5))
dendrogram(Z, leaf_font_size=14)
plt.show()
```

```python
Z = linkage(X, 'complete')
plt.figure(figsize=(12, 5))
dendrogram(Z, leaf_font_size=14)
plt.show()
```

```python
Z = linkage(X, 'centroid')
plt.figure(figsize=(12, 5))
dendrogram(Z, leaf_font_size=14)
plt.show()
```

```python
Z = linkage(numericalDF, 'centroid')
plt.figure(figsize=(12, 5))
dendrogram(Z, leaf_font_size=14)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
dendrogram(
    Z,
    truncate_mode='lastp',  # mostrar slo los ltims p clusters
    p=15,                   
    show_leaf_counts=True,  # mostrar entre parntesis el nmero de elementos en cada cluster
    leaf_font_size=14)
plt.show()
```
