```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
df = pd.read_csv('movie_metadata1.csv')
del df['Unnamed: 0']
del df['Unnamed: 0.1']
print(len(df))
print(list(df))
df.head()
```

```python
dicc_directores = dict()
count = 0
array = []
for i in df['director_name']:
    if i not in dicc_directores.keys():
        count = count + 1
        dicc_directores.update({i:count})
    array.append(dicc_directores[i])
#print(array)
#print(dicc_directores)
df['director_name'] = array
```

```python
array = []
a = df['num_critic_for_reviews'][4]
for i in df['num_critic_for_reviews']:
    if np.isnan(i):
        i = 0
    array.append(i)
df['num_critic_for_reviews'] = array
```

```python
array = []
a = df['duration'][4]
for i in df['duration']:
    if np.isnan(i):
        i = 0
    array.append(i)
df['duration'] = array
```

```python
array = []
a = df['gross'][4]
for i in df['gross']:
    if np.isnan(i):
        i = 0
    array.append(i)
df['gross'] = array
```

```python
dicc_genres = dict()
count = 0
array = []
print(df['genres'][0].split('|'))
for values in df['genres']:
    a = []
    for i in values.split('|'):
        if i not in dicc_genres.keys():
            count = count + 1
            dicc_genres.update({i:count})
        a.append(dicc_genres[i])
    array.append(a)
df['genres'] = array
```

```python
dicc_actores1 = dict()
count = 0
array = []
for i in df['actor_1_name']:
    if i not in dicc_actores1.keys():
        count = count + 1
        dicc_actores1.update({i:count})
    array.append(dicc_actores1[i])
#print(array)
#print(dicc_actores1)
df['actor_1_name'] = array
```

```python
#movie_title
dicc_titulos = dict()
count = 0
array = []
print(len(df['movie_title']))
for i in df['movie_title']:
    if i not in dicc_titulos:
        count = count + 1
        dicc_titulos.update({i:count})
    array.append(dicc_titulos[i])
print(len(array))
#print(array)
print(np.unique(array,return_index=True))
print(len(dicc_titulos))
dupes = [x for n, x in enumerate(array) if x in array[:n]]
print((set(dupes)))
```
